# R-sq Demo

Generate some random data, fit a regression line, and display the corresponding R-sq value

```python
# Example of Linear Regression with some plots
# let's create some random data

import numpy as np

POINTS = 2
X = 2 * np.random.rand(POINTS, 1)
y = 1 + 1/3 * X + np.random.randn(POINTS, 1)

import matplotlib.pyplot as plt
#plt.scatter(X, y, alpha=0.3)

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print('int: %.2f  coef: %.2f' %(lin_reg.intercept_[0], lin_reg.coef_[0][0]))


f = lambda x:lin_reg.intercept_[0] + x*lin_reg.coef_[0][0]
x = np.array([min(X), max(X)])
plt.plot(x, f(x), "r") 
plt.scatter(X,y, alpha=0.3, color='green')

R2_score = lin_reg.score(X, y)
print('R2: %5.3f\n' %R2_score)
if R2_score > 0.5:
    print("*****"*5)
if R2_score > 0.75:
    print("*****"*7)
if R2_score > 0.90:
    print("*****"*10)
```
